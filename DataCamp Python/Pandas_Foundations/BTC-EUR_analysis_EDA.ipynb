```python
# Time-series analysis BTC-EUR

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

BTC_EUR_df = pd.read_csv('BTC-EUR.csv', index_col=0, parse_dates=True)

print(BTC_EUR_df)

print("\nINFO:")
print(BTC_EUR_df.info())

print("\nCOLONNE:")
print(BTC_EUR_df.columns)

print("INDEX:")
print(BTC_EUR_df.index)

print("\n\nSTATISTICHE DESCRITTIVE:\n")
print(BTC_EUR_df.describe)

print("\n\nSTATISTICHE DESCRITTIVE(solo dei prezzi di apertura e chiusura):\n")
BTC_EUR_sub = BTC_EUR_df[['Open', 'Close']]
print(BTC_EUR_sub.describe())
```

```python
# pandas line chart
# plot == 'line chart'

BTC_EUR_df.plot(y=['Open', 'Close'])
plt.show()


#  You'll specify subplots=True inside .plot() to generate two separate plots.

print("\n\nSUBPLOT:")

BTC_EUR_df.plot(y=['Open', 'Close'], subplots=True)
plt.show()
```

```python
# pandas scatter plots

"""
Pandas scatter plots are generated using the kind='scatter' keyword argument. 
Scatter plots require that the x and y columns be chosen by specifying the x and y parameters inside .plot(). 
Scatter plots also take an s keyword argument(s=size) to provide the radius of each circle to plot in pixels.
"""

# Generate a scatter plot
BTC_EUR_df.plot(kind='scatter', x='Open', y='Close')           # 's=sizes' if you wanna spessore of the puntini

# Add the title
plt.title('Open-Close of \'BTC-EUR\' times series')

# Add the x-axis label
plt.xlabel('Open price')

# Add the y-axis label
plt.ylabel('Close price')

# Display the plot
plt.show()
```

```python
# pandas box plots      kind='box'
# BOX PLOTS are a great way to visualize important summary statistics.

# Make a list of the column names to be plotted: cols
cols = ['Open', 'Close']

# Generate the box plots
BTC_EUR_df[cols].plot(kind='box', subplots=True)

# Display the plot
plt.show()

print("\n\nSenza subplots:")

BTC_EUR_df[cols].plot(kind='box')         # senza subplots, stavolta
plt.show()
```

```python
# pandas hist
# PDF                       normed/density=True
# CDF                       cumulative=True

"""
Pandas relies on the .hist() method to not only generate histograms, 
but also plots of probability density functions (PDFs) and cumulative density functions (CDFs).
"""

# The 'normed' kwarg was deprecated in Matplotlib 2.1 and will be removed in 3.1. Use 'density' instead.
# alternative="'density'", removal="3.1")

BTC_EUR_df['Open'].plot(kind='hist', bins=10, density=True)
plt.show()

BTC_EUR_df['Open'].plot(kind='hist', bins=10, density=True, cumulative=True)
plt.show()
```
